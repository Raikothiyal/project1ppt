```python
columns = ['season','mnth','hr','holiday','weekday','weathersit','temp','hum','windspeed','casual','registered','cnt','atemp']
weather1[columns]= weather1[columns].apply(pd.to_numeric)
```

```python
Q1= weather1['registered'].quantile(.25)
Q3= weather1['registered'].quantile(.75)
```

```python
def find_outliers(df, columns):
    Q1= df[columns].quantile(.25)
    Q3= df[columns].quantile(.75)
    iqr = Q3-Q1
    min_r = Q1 - 1.5*iqr
    max_r = Q3 + 1.5*iqr
    outliers_indices = df.index[(df[columns] < min_r) | (df[columns]> max_r)]
    return outliers_indices
```

```python
find_outliers (weather1,'temp')
```

```python
find_outliers (weather1, 'windspeed')
```

```python
find_outliers (weather1, 'casual')
```

```python
find_outliers (weather1, 'cnt')
```
